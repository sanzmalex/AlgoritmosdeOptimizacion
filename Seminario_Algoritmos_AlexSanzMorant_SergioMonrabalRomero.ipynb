# Diseña un algoritmo para resolver el problema por fuerza bruta

**Respuesta**

```python
#Fuerza Bruta

#Se definen dos listas con los números y las operaciones
lista_operaciones = ["+","-","*","/"]
lista_numeros = [1,2,3,4,5,6,7,8,9]

def operaciones_paraFB(target):
    nums = lista_numeros
    ops = lista_operaciones

    #Iteramos sobre la lista de números tomando 5 números distintos para comprobar
    for num_uno in nums:
        for num_dos in nums:
            if num_dos == num_uno:
                continue
            for num_tres in nums:
                if num_tres == num_uno or num_tres == num_dos:
                    continue
                for num_cuatro in nums:
                    if num_cuatro == num_uno or num_cuatro == num_dos or num_cuatro == num_tres:
                        continue
                    for num_cinco in nums:
                        if (num_cinco == num_uno or num_cinco == num_dos or
                            num_cinco == num_tres or num_cinco == num_cuatro):
                            continue

                        #Iteramos sobre la lista de operaciones tomando 4 distintas para comprobar
                        for op_uno in ops:
                            for op_dos in ops:
                                if op_dos == op_uno:
                                    continue
                                for op_tres in ops:
                                    if op_tres == op_uno or op_tres == op_dos:
                                        continue
                                    for op_cuatro in ops:
                                        if op_cuatro == op_uno or op_cuatro == op_dos or op_cuatro == op_tres:
                                            continue

                                        #Evaluamos si la expresión con las operaciones calcula el número que queremos, si no continuamos iterando
                                        expr = f"{num_uno}{op_uno}{num_dos}{op_dos}{num_tres}{op_tres}{num_cuatro}{op_cuatro}{num_cinco}"
                                        if eval(expr) == target:
                                            return expr
    return None #Si no encontramos ninguna combinación devolvemos None

for s in range(-70, 80):
    print(f"{s}: {operaciones_paraFB(s)}")
```

# Calcula la complejidad del algoritmo por fuerza bruta

**Respuesta**

De la discusión anterior se deduce que el algoritmo tiene 9 bucles for anidados: 5 recorren la lista de números, de tamaño N, y los otros 4 recorren la lista de operaciones. Como el número de operaciones que se puede pasar está acotado, ese factor se considera constante. Además, el resto de instrucciones de la función son de orden constante, por lo que no afectan al orden de crecimiento.
En consecuencia, la complejidad temporal del algoritmo es 𝓞($N^5$).

# (*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta.

**Respuesta**

```python
#Se definen dos listas con los números y las operaciones
lista_operaciones=["+","-","*","/"]
lista_numeros=[1,2,3,4,5,6,7,8,9]

def operaciones_para(target):
  #Se crea una copia de las listas anteriores
  temp_ops = lista_operaciones[:]
  temp_nums = lista_numeros[:]

  def buscarDigito(stage, actualExpr):

    #Pequeña poda si ya es imposible llegar al valor cuando queda solo "+" y "-"
    if stage > 0:
      solo_mas_menos = True
      for op in temp_ops: #M operaciones
        if op != "+" and op != "-":#Operaciones constantes
          solo_mas_menos = False
          break
      if solo_mas_menos:
        actualVal = eval(actualExpr)
        s = 0
        for x in temp_nums:#N operaciones
          s += x
        if target < actualVal - s or target > actualVal + s:#Operaciones constantes
          return None

    #Si estamos realizando el último paso evaluamos la expresión
    if stage == 5: #Operaciones constantes
      if eval(actualExpr) == target:
        return actualExpr
      return None

    #Se aplica Backtracking para obtener los números y las operaciones
    for i in range(len(temp_nums)): #N operaciones
      actual_dig = temp_nums.pop(i)
      if stage == 0:
          ans = buscarDigito(stage + 1, str(actual_dig))
      else:
        ans = None
        for j in range(len(temp_ops)): #M operaciones
          actual_op = temp_ops.pop(j) #Operaciones constantes

          ans = buscarDigito(stage + 1, actualExpr + actual_op + str(actual_dig))

          temp_ops.insert(j, actual_op) #Operaciones constantes
          if ans is not None: #Operaciones constantes
            break

      temp_nums.insert(i, actual_dig) #Operaciones constantes
      if ans is not None: #Operaciones constantes
        return ans

    return None

  return buscarDigito(0, "")

for s in range(20):
  print(str(s)+": ",operaciones_para(s))
```

# Argumenta porque crees que el algoritmo anterior mejora el algoritmo por fuerza bruta

Pese a que la heurística empeora, aparentemente, el orden de complejidad del algoritmo, pasando de 𝓞($N^{5}$) a 𝓞($N^{6}$) el algoritmo de Backtracking, el cual tiene mayor coste asintótico que el de Fuerza Bruta, que tiene 𝓞($N^{5}$); en la práctica el algoritmo de Backtracking será más eficiente que el de Fuerza Bruta. Esto es debido a que la poda es capaz de evitar que se recorra todo el árbol en la mitad de los casos (ya que de las 24 combinaciones posibles de operaciones, la mitad acaban en suma o resta), siendo una cuarta parte de los casos totales especialmente eficientes (6 combinaciones en total) dado que acaban en suma y resta (o viceversa).

Esto mismo podemos comprobarlo si hacemos un estudio empírico de nuestro problema. Veamos primero que el algoritmo de Backtracking más rápido que el de Fuerza Bruta en una mayor cantidad de casos:

```python
import time

lista_numeros=[1,2,3,4,5,6,7,8,9]
mejorFB = []
mejorBT = []
igualDeRapidos = []

for s in range(-69,78):
  startFB = time.perf_counter()
  operaciones_paraFB(s)
  stopFB = time.perf_counter()
  tiempoFB= stopFB-startFB

  startBT = time.perf_counter()
  operaciones_para(s)
  stopBT = time.perf_counter()
  tiempoBT= stopBT-startBT

  if tiempoFB > tiempoBT:
    mejorBT.append(s)
  elif tiempoFB < tiempoBT:
    mejorFB.append(s)
  else:
    igualDeRapidos.append(s)

print("Hay ", len(mejorFB)," números para los cuales el algoritmo de Fuerza Bruta es más rápido, y son los siguientes: ", mejorFB)
print("Hay ", len(mejorBT)," números para los cuales el algoritmo de Backtracking es más rápido, y son los siguientes: ", mejorBT)
print("Hay ", len(igualDeRapidos)," números para los cuales los algoritmo de Backtracking y Fuerza Bruta son igual de rápidos, y son los siguientes: ", igualDeRapidos)
```

Cabría preguntarse también cuál es el tiempo medio que tarda el algoritmo de Backtracking en obtener el número que busca, y si es más rápido en promedio que el algoritmo de Fuerza Bruta. Veamos esto empíricamente:

```python
timeFB=[]
timeBT=[]
for s in range(-69,78):
  startFB = time.perf_counter()
  operaciones_paraFB(s)
  stopFB = time.perf_counter()
  tiempoFB= stopFB-startFB
  timeFB.append(tiempoFB)

  startBT = time.perf_counter()
  operaciones_para(s)
  stopBT = time.perf_counter()
  tiempoBT= stopBT-startBT
  timeBT.append(tiempoBT)

tiempoMedioFB = sum(timeFB)/len(timeFB)
tiempoMedioBT = sum(timeBT)/len(timeBT)

if tiempoMedioFB > tiempoMedioBT:
  print(f"El algoritmo de Backtracking con {tiempoMedioBT}s es más rápido en promedio que el de Fuerza Bruta con {tiempoMedioFB}s.")
elif tiempoMedioFB < tiempoMedioBT:
  print(f"El algoritmo de Fuerza Bruta con {tiempoMedioFB}s es más rápido en promedio que el de Backtracking con {tiempoMedioBT}s.")
else:
  print("Son igual de rápidos.")
print(f"Diferencia en el tiempo promedio: {tiempoMedioFB/tiempoMedioBT}")
```

Creamos una función que comprueba los tiempos:

```python
def comprobacionTiempos():

  #Declaramos la variables que vamos a usar y nos traemos la lista de números que son capaces de calcular los algoritmos
  lista_comprobacion=rangoAlgs()
  mejorFB = []
  mejorBT = []
  igualDeRapidos = []
  timeFB=[]
  timeBT=[]

  #Recorremos lista_comprobación y vamos obteniendo los tiempos y la lista de tiempos
  for s in lista_comprobacion:
    startFB = time.perf_counter()
    operaciones_paraFB(s)
    stopFB = time.perf_counter()
    tiempoFB= stopFB-startFB
    timeFB.append(tiempoFB)

    startBT = time.perf_counter()
    operaciones_para(s)
    stopBT = time.perf_counter()
    tiempoBT= stopBT-startBT
    timeBT.append(tiempoBT)

    #Obtenemos las listas de los números con los que cada algortimo es más rápido
    if tiempoFB > tiempoBT:
      mejorBT.append(s)
    elif tiempoFB < tiempoBT:
      mejorFB.append(s)
    else:
      igualDeRapidos.append(s)

  #Obtenemos los tiempos medios y mostramos los números con los que cada algoritmo es más rápido y los tiempos
  tiempoMedioFB = sum(timeFB)/len(timeFB)
  tiempoMedioBT = sum(timeBT)/len(timeBT)
  print("Hay ", len(mejorFB)," números para los cuales el algoritmo de Fuerza Bruta es más rápido, y son los siguientes: ", mejorFB)
  print("Hay ", len(mejorBT)," números para los cuales el algoritmo de Backtracking es más rápido, y son los siguientes: ", mejorBT)
  print("Hay ", len(igualDeRapidos)," números para los cuales los algoritmo de Backtracking y Fuerza Bruta son igual de rápidos, y son los siguientes: ", igualDeRapidos)

  if tiempoMedioFB > tiempoMedioBT:
    print(f"El algoritmo de Backtracking con {tiempoMedioBT}s es más rápido en promedio que el de Fuerza Bruta con {tiempoMedioFB}s.")
  elif tiempoMedioFB < tiempoMedioBT:
    print(f"El algoritmo de Fuerza Bruta con {tiempoMedioFB}s es más rápido en promedio que el de Backtracking con {tiempoMedioBT}s.")
  else:
    print("Son igual de rápidos.")
  print(f"Diferencia en el tiempo promedio: {tiempoMedioFB/tiempoMedioBT}")

  return None
```
